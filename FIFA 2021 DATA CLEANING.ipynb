# Contract

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == "Free" or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year         
    return start_date, end_date, contract_length

# Apply the function to contract column and create new columns 
new_cols = ['Contract Start', 'Contract End', 'Contract length(year)'] 
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

# Loop through column names and insert date 
for i in range(len(new_cols)):
    new_col_name = new_cols[i]
    if new_col_name not in fifa.columns:
        fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_col_name, value=new_data[i])
```

```python
# added three new columns; Contract Start, Contract End, Contract lenght
fifa.sample(5)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract length(year)']].sample(5)
```

```python
# Contract categories
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# Add contract status column
fifa.insert(fifa.columns.get_loc('Contract length(year)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.sample(5)
```

```python
fifa[['Contract','Contract Start', 'Contract End','Contract length(year)','Contract Status']].sample(10)
```

# Height

```python
# players height in centimeters
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
# Convert the height into centimeter and into integers

def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

#Apply the function to height column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
# Rename the column
fifa = fifa.rename(columns = {'Height': 'Height(cm)'})
fifa.sample(5)
```

# Weight

```python
# players weight in Kilograms
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)

# Applying the function to the weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
# Rename the column
fifa = fifa.rename(columns = {'Weight': 'Weight(kg)'})
fifa.sample(5)
```
